```python
class LoginError(Exception):
  pass
crct_psw="5627"
attempts=3
while attempts>0:
  try:
    psw=input("Enter the password:")
    if psw!=crct_psw:
      attempts-=1
      raise LoginError("wrong password")
    print("Login successful")
    break
  except LoginError as e:
     print(e)
else:
  print("Account Locked")
```

```python
balance=2000
try:
  amount=float(input("Enter the amount:"))
  if amount>balance:
    raise Exception("Insufficient balance")
  balance-=amount
  print("Withdrawal succesful")
  print("Remaining Balance:",balance)
except Exception as e:
  print(e)
```

```python
class AgeError(Exception):
  pass
try:
  age=int(input("Enter the age:"))
  if age < 0:
    raise AgeError("The age cannot be negative")
  elif age < 18:
    raise AgeError("Must be 18 or older")
  else:
    print("Valid Age")
except AgeError as e:
  print(e)
except ValueError:
  print("Invalid input")
```

```python
f = open("marks.txt", "w")
total = 0
for i in range(3):
    name = input("Enter student name: ")
    mark = int(input("Enter mark: "))
    total = total + mark
    f.write(name + " : " + str(mark) + "\n")
average = total / 3
f.write("Average : " + str(average))
f = open("marks.txt", "r")
print(f.read())
f.close()
```

```python
f = open("students.txt", "w")
f.write("Hermi,90\n")
f.write("Santrhu,92\n")
f.write("James,95\n")
f = open("students.txt", "r")
lines = f.readlines()
name = input("Enter student name: ")
new_mark = input("Enter new mark: ")
f = open("students.txt", "w")
for line in lines:
   data = line.strip().split(",")
   if data[0] == name:
       f.write(name + "," + new_mark + "\n")
   else:
        f.write(line)
f = open("students.txt", "r")
print(f.read())
f.close()
```
